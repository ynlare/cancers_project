```python
import pandas as pd 

import matplotlib.pyplot as plt 

import seaborn as sns 
```

```python
## Import dataset cancer_b

cancerb = pd.read_csv('/Users/Yedou/Desktop/HistgramCancer/cancer_b.csv')

cancerb.head()
```

```python

# change Index cancer_b to id

cancerb.set_index('Id', inplace = True)
```

```python
cancerb.head() # The first five lines
```

```python
## Import dataset cancer_m

cancerm = pd.read_csv('/Users/Yedou/Desktop/HistgramCancer/cancer_m.csv')

cancerm.head()
```

```python
cancerb.info()
```

```python
cancerm.info()
```

```python
# # change Index cancer_m to id
cancerm.set_index('Id', inplace = True)
```

```python
cancerm.head() # The first five lines
```

```python

# Histogram of cancerb and cancerm on the same figure 
#  mean representation 

plt.figure( figsize = (10,5))

sns.distplot(a = cancerb['Area (mean)'], label = 'Benigne',kde = False)

sns.distplot(a = cancerm['Area (mean)'], label = 'Maligne',kde = False)

plt.legend()

plt.show()
```

```python
# worst representation

plt.figure( figsize = (10,5))

sns.distplot(a = cancerb['Area (worst)'], label = 'Benigne',kde = False)

sns.distplot(a = cancerm['Area (worst)'], label = 'Maligne',kde = False)

plt.legend()

plt.show()
```

```python
#  mean kde Plot

plt.figure( figsize = (10,6))

sns.kdeplot(data = cancerb['Area (mean)'], label = 'Benigne',shade = True)

sns.kdeplot(data = cancerm['Area (mean)'], label = 'Maligne',shade = True)

plt.legend()

plt.show()
```

```python
#  worst kde Plot

plt.figure( figsize = (10,6))

sns.kdeplot(data = cancerb['Area (worst)'], label = 'Benigne',shade = True)

sns.kdeplot(data = cancerm['Area (worst)'], label = 'Maligne',shade = True)

plt.legend()

plt.show()
```
